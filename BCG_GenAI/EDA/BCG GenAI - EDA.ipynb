## Importing Packages and EDA

```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data
```

### Analysing Trends

Get the % year on Growth of the Companies

```python
data['Revenue Growth (%)'] = data.groupby(['Company'])['Total Revenue'].pct_change() * 100
data['Operating Expenses Growth (%)'] = data.groupby(['Company'])['Total Operating Expenses'].pct_change() * 100
data['Net Income Growth (%)'] = data.groupby(['Company'])['Net Income'].pct_change() * 100
data['Total Assets Growth (%)'] = data.groupby(['Company'])['Total Assets'].pct_change() * 100
data['Total Liabilities Growth (%)'] = data.groupby(['Company'])['Total Liabilities'].pct_change() * 100
data['Operating Activities Cash Growth (%)'] = data.groupby(['Company'])['Cash From Operating Activities'].pct_change() * 100
```

```python
data
```

### Removing Null Values

```python
data.isnull().sum()
```

fill the missing data with 0 as you don't have 2020's data to calculate the % Growth increase for the year 2021

```python
data = data.fillna(0)
```

### Create new dataset

```python
data.to_csv('final_data.csv')
```

```python
summary = data.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Operating Expenses Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Total Assets Growth (%)': 'mean',
    'Total Liabilities Growth (%)': 'mean',
    'Operating Activities Cash Growth (%)': 'mean'
}).reset_index()

print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```

```python
summary.to_csv('summary_final_data.csv')
```
